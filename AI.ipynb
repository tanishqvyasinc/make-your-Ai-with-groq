```python
import os
!pip install langchain-groq # Ensure the package is installed within this cell's context
from langchain_groq import ChatGroq

# 🔑 Set your Groq API key from groq wbebsite
os.environ["GROQ_API_KEY"] = "API key here" #put your api key here

# 🤖 Initialize model also go for lastest model form groq webside
llm = ChatGroq(
    model="llama-3.1-8b-instant",   # Use a currently supported model
    temperature=0.72
)

# 🔁 Simple text generation loop
while True:
    user_input = input("\nYou: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Goodbye!")
        break

    response = llm.invoke(user_input)

    print("\nAI:", response.content)
```
